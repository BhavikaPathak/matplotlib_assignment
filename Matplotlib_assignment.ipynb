```python
#                                          ANSWER 2 
# The scatter (previously called XY) plots while comparing various data variables to determine the connection between dependent and independent variables. The data gets expressed as a collection of points clustered together meaningfully.
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This Is My X Axis")
plt.ylabel("This Is My Y Axis")
plt.title("Scatter Plot")
```

```python
#                                        ANSWER 3
# Matplotlibs pyplot module has a convenience function called Matplotlib subplots() that is used to create multiple plots on a figure with a single call. 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot()
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot()
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot()
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot()
plt.plot(x,y)
```

```python
#                                       ANSWER 4 
# Bar plots are a type of data visualization used to represent data in the form of rectangular bars.The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.
# The Matplotlib library in Python is widely used to create bar plots. The bar() function in Matplotlib is used to create bar plots; and accepts data in the form of lists, numpy arrays, and pandas dataframes.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit,color=['#9C081C','#08359C','#9C088C','#9C8F08'])
plt.xlabel("This Is Profit data")
plt.ylabel("This Is Company data")
plt.title(" Horizontal Bar Plot")
```

```python
#                                                        ANSWER 5
#A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum.
#Boxplots are a measure of how well distributed the data in a data set is.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box = [ box1 , box2 ]
plt.boxplot(box)
```
